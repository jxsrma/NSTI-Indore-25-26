```python
import pandas as pd
import numpy as np
from faker import Faker

# Initialize faker
fake = Faker()
np.random.seed(42)

# Number of rows
n_rows = 5000

# Create base data
data = {
    "EmployeeID": range(1001, 1001 + n_rows),
    "Name": [fake.name() for _ in range(n_rows)],
    "Department": np.random.choice(
        ["Sales", "HR", "IT", "Finance", "Marketing", None],
        size=n_rows,
        p=[0.2, 0.15, 0.25, 0.15, 0.15, 0.1]
    ),
    "Age": np.random.choice(
        list(range(22, 60)) + [None],
        size=n_rows,
        p=[0.97 / 38] * 38 + [0.03]
    ),
    "Salary": np.random.choice(
        list(range(35000, 120000, 1000)) + [None],
        size=n_rows,
        p=[0.96 / 85] * 85 + [0.04]
    ),
    "JoinDate": [
        fake.date_between(start_date="-10y", end_date="today")
        if np.random.rand() > 0.05 else None
        for _ in range(n_rows)
    ],
    "Email": [
        fake.email() if np.random.rand() > 0.1 else None
        for _ in range(n_rows)
    ],
    "ManagerID": np.random.choice(
        list(range(2001, 2050)) + [None],
        size=n_rows,
        p=[0.95 / 49] * 49 + [0.05]
    )
}

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("large_employees_dataset.csv", index=False)

# Preview
print(df.head())
print("\nMissing values per column:\n")
print(df.isna().sum())
```
