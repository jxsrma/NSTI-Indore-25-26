```python
import pandas as pd

data_easy_encode = {
    "Water_Intake": [
        "Low", "Medium", "High", "Medium", "Low",
        "High", "Medium", "Low", "High", "Medium"
    ],
    
    "Workout_Regularity": [
        "Never", "Sometimes", "Daily", "Sometimes", "Never",
        "Daily", "Sometimes", "Never", "Daily", "Sometimes"
    ],
    
    "Diet_Type": [
        "Junk", "Balanced", "Balanced", "Junk", "Junk",
        "Balanced", "Balanced", "Junk", "Balanced", "Balanced"
    ],
    
    "Sleep_Quality": [
        "Poor", "Average", "Good", "Average", "Poor",
        "Good", "Average", "Poor", "Good", "Average"
    ],
    
    "Uses_Fitness_App": [
        "Yes", "No", "Yes", "No", "No",
        "Yes", "No", "Yes", "Yes", "No"
    ],
    
    # Numeric target
    "Health_Score": [40, 65, 85, 60, 45, 90, 70, 42, 88, 68]
}

df = pd.DataFrame(data_easy_encode)
df
```

```python
def changeToLabel(str):
    if str == 'Low':
        return 0
    elif str == 'Medium':
        return 1
    else:
        return 2
```

```python
df['Water_Intake'] = df['Water_Intake'].apply(lambda x: changeToLabel(x))
df
```

```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
```

```python

df['Workout_Regularity'] = LE.fit_transform(df['Workout_Regularity'])
df
```

```python
WorkOutClasses = LE.classes_
WorkOutClasses
```

```python

df['Sleep_Quality'] = LE.fit_transform(df['Sleep_Quality'])
df
```

```python
SleepQualityClasses = LE.classes_
SleepQualityClasses
```

```python
OneHot = pd.get_dummies(df[['Diet_Type','Uses_Fitness_App']])
OneHot
```

```python
FinalDF = pd.concat([df,OneHot],axis=1)
FinalDF
```

```python
dataset = FinalDF.select_dtypes(['number','boolean'])
dataset
```

```python
dataset['Health_Score'] = dataset['Health_Score'] * 2
dataset
```
