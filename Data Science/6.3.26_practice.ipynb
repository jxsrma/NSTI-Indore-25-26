```python
import matplotlib.pyplot as plt
```

```python
CountOfHouseByYear = df.groupby('Year_Built').count()['Years Old']
```

```python
CountOfHouseByYear
```

```python
plt.plot(CountOfHouseByYear.index,CountOfHouseByYear.values)
plt.xlabel('Year')
plt.ylabel('Count of Houses Build')
plt.title('Count of Houses Per Year')
plt.show()
```

```python
plt.bar(CountOfHouseByYear.index,CountOfHouseByYear.values)
plt.xlabel('Year')
plt.ylabel('Count of Houses Build')
plt.title('Count of Houses Per Year')
plt.show()
```

```python
HouseCountByQualityNeigh = df.groupby('Neighborhood_Quality').count()['Square_Footage']
HouseCountByQualityNeigh
```

```python
plt.bar(HouseCountByQualityNeigh.index,HouseCountByQualityNeigh.values)
plt.xlabel('Quality')
plt.xticks(HouseCountByQualityNeigh.index)
plt.ylabel('Count of Houses')
plt.title('Count of Houses By Quality Neighborhood')
plt.show()
```

```python
plt.pie(HouseCountByQualityNeigh.values,labels=HouseCountByQualityNeigh.index,autopct='%1.1f%%')
plt.show()
```
