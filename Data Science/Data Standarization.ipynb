```python
import numpy as np
data = np.random.randint(low=1,high=20,size=30)
data = list(data)
data
```

```python
minimun = min(data)
maximum = max(data)
```

```python
maximum
```

```python
minimun
```

```python
def normalization(n,max,min):
    range = max - min
    return (n-min)/range
```

```python
normalization(200,300,100)
```

```python
normalForm = []
for i in data:
    normalForm.append(normalization(i,maximum,minimun))
normalForm
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
import pandas as pd
s1 = pd.Series(data)
data
```

```python
stdSclr = StandardScaler()

scaledData = stdSclr.fit_transform(s1.values.reshape(-1, 1))
scaledData
```

```python
scaledList = []
for i in scaledData:
    scaledList.append(i[0])
```
