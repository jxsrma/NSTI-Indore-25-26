```python
import pandas as pd

data_viz = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun", "Mon", "Tue", "Wed"],
    "Steps": [3000, 4500, 5000, 7000, 8000, 10000, 9000, 3500, 4800, 5200],
    "Sleep_Hours": [6, 6.5, 7, 7.5, 8, 8.5, 8, 6, 6.5, 7],
    "Study_Hours": [1, 2, 2.5, 3, 3.5, 1, 0.5, 1.5, 2, 2.5],
    "Energy_Level": [4, 5, 5, 6, 7, 6, 6, 4, 5, 5],
    "Junk_Food": ["Yes", "No", "No", "No", "No", "Yes", "Yes", "Yes", "No", "No"],
    "Mood": ["Low", "Medium", "Medium", "Good", "Good", "Medium", "Medium", "Low", "Medium", "Medium"]
}
df = pd.DataFrame(data_viz)
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(
    df['Day'],
    df['Steps'],

)
```

```python
plt.plot(
    df['Day'][:7],
    df['Steps'][:7],
    color = 'Black',
    marker = '*'
)
plt.xlabel('Days of Week')
plt.ylabel('Steps')
plt.title("Line Plot â€“ Daily Steps Trend")
```

```python
averageStepsPerDay = df.groupby('Day')['Steps'].mean()
averageStepsPerDay
```

```python
plt.plot(
    averageStepsPerDay.index,
    averageStepsPerDay.values,
    color = 'Black',
    marker = '*'
)
plt.xlabel('Days of Week')
plt.ylabel('Steps')
plt.title("Line Plot â€“ Average Steps Trend")
```

```python
# plt.plot(
#     kind = 'Bar',
#     x = averageStepsPerDay.index,
#     y = averageStepsPerDay.values,
#     color = 'Black',
#     marker = '*'
# )
# plt.xlabel('Days of Week')
# plt.ylabel('Steps')
# plt.title("Line Plot â€“ Daily Steps Trend")

averageStepsPerDay.plot(kind='bar',x=averageStepsPerDay.index,y=averageStepsPerDay.values)
plt.xlabel('Days of Week')
plt.ylabel('Steps')
plt.title("Bar Plot â€“ Average Steps Trend")
plt.grid(True)
```

```python
# Sleep_Hours vs Energy_Level
plt.figure(figsize=(10,4))
plt.scatter(df['Sleep_Hours'],df['Energy_Level'])
plt.xlabel('Sleep Hours')
plt.ylabel('Energy Level')
plt.title("Scatter â€“ Sleep Hours vs Energy Level")
```
