# Text Preprocessing with NLTK

STEP 1: Install & Import NLTK

````python
import nltk
````

STEP 2: Input Sentence

````python
sentance = "Microsoft Corporation is an American multinational technology company headquartered in Redmond, Washington. The company became influential in the rise of personal computers through software like Windows and has since expanded into areas such as Internet services, cloud computing, artificial intelligence, video gaming, and more. A Big Tech company, Microsoft is the largest software company by revenue, one of the most valuable public companies, and one of the most valuable brands globally."
````

STEP 3: Lower Case the sentenace

````python
sentance = sentance.lower()
sentance
````

STEP 4: Tokenization
- Sentance into Words
- with the help of word_tokenize function from nltk.tokenize

````python
from nltk.tokenize import word_tokenize
````

````python
nltk.download('punkt')
nltk.download('punkt_tab')
````

````python
token = word_tokenize(sentance)
token
````

STEP 5: Stopword Removal
- We have to remove words like is, and, for, not which add little meaning.

- To find all stopwords, we have to import stopwords from nltk.corpus
- Corpus = Large dataset of words

````python
from nltk.corpus import stopwords
````

````python
stopwordList = stopwords.words('english')
````

````python
token = [i for i in token if i not in stopwordList]
````

````python
print(token)
````

STEP 5: Remove Symbols
- using isalnum()

````python
token = [i for i in token if i.isalnum()]
````

STEP 5: Stemming
- To cut down the words to their root form by using PorterStemmer from nltk.stem

````python
from nltk.stem import PorterStemmer
````

````python
ps = PorterStemmer()
````

````python
ps.stem('leaves')
````

````python
stemmedTokens = []
for i in token:
    stemmedTokens.append(ps.stem(i))
stemmedTokens
````

STEP 6: Lemmatization
- We can also use Lemmatization as it is better than stemming
- Using WordNetLemmatizer from nltk.stem

````python
from nltk.stem import WordNetLemmatizer
````

````python
wnl = WordNetLemmatizer()
````

````python
wnl.lemmatize("reading",pos='v')
````

STEP 6: Apply POS Taggin
- with nltk.pos_tag()

````python
tags = nltk.pos_tag(token)
tags
````

STEP 7: Named Entity Recognition
- Using nltk.ne_chunk(pos_tags)

````python
tree = nltk.ne_chunk(tags)
tree
````

````python
import spacy

# Load a pre-trained model
nlp = spacy.load("en_core_web_sm")

# Process text
doc = nlp("Steve Job's Apple is looking at buying U.K. startup for $1 billion")

# Iterate through entities
for ent in doc.ents:
    print(ent.text, ent.label_)
# Output: Apple (ORG), U.K. (GPE), $1 billion (MONEY)
````
