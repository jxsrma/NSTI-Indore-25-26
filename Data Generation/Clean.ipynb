```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import BayesianRidge

df = pd.read_csv('D:\\EduNet\\NSTI_Indore25_26\\ML\\Datasets\\city_day.csv')

# Drop non-useful columns
df = df.drop(columns=['City','AQI_Bucket'], errors='ignore')

# Convert Date
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df['Date'] = df['Date'].map(pd.Timestamp.toordinal)

# Remove bad columns
df = df.loc[:, df.isnull().mean() < 0.5]
df = df.loc[:, df.nunique() > 1]

# Imputer (FAST)
imputer = IterativeImputer(
    estimator=BayesianRidge(),
    max_iter=5,
    random_state=42
)

df_imputed = pd.DataFrame(
    imputer.fit_transform(df),
    columns=df.columns
)

# Save
df_imputed.to_csv('D:\\EduNet\\NSTI_Indore25_26\\ML\\Datasets\\cleaned_aqi_data.csv', index=False)

print("✅ Done! Clean dataset saved.")
```
