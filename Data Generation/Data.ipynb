```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Number of rows
n = 50

# Generate date range
dates = pd.date_range(start='2024-01-01', periods=n, freq='D')

# Generate realistic temperature data
base_temp = 25
trend = np.linspace(0, 5, n)  # slight upward trend
seasonal = 3 * np.sin(np.linspace(0, 3*np.pi, n))  # seasonal variation
noise = np.random.normal(0, 1, n)  # random noise

temperature = base_temp + trend + seasonal + noise

# Create DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Temperature': np.round(temperature, 2)
})

# Save to CSV (optional)
df.to_csv("D:\\EduNet\\NSTI_Indore25_26\\Deep Learning\\Dataset\\temperature_data.csv", index=False)

print(df.head(10))
```
