```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 200

# ---------------------
# CATS (Overlapping)
# ---------------------
cat_height = np.random.normal(loc=30, scale=5, size=n)   # cm
cat_weight = np.random.normal(loc=4, scale=1.2, size=n)  # kg

cats = pd.DataFrame({
    "Height_cm": cat_height.round(2),
    "Weight_kg": cat_weight.round(2),
    "Animal": "Cat"
})

# ---------------------
# DOGS (Overlapping)
# ---------------------
dog_height = np.random.normal(loc=40, scale=10, size=n)   # cm
dog_weight = np.random.normal(loc=10, scale=4, size=n)    # kg

dogs = pd.DataFrame({
    "Height_cm": dog_height.round(2),
    "Weight_kg": dog_weight.round(2),
    "Animal": "Dog"
})

# ---------------------
# Combine & Shuffle
# ---------------------
df = pd.concat([cats, dogs], ignore_index=True)
df = df.sample(frac=1).reset_index(drop=True)

# Add random noise feature (to confuse model slightly)
df["Tail_Length_cm"] = np.random.normal(30, 8, len(df)).round(2)

# Save CSV
df.to_csv("D:\\EduNet\\NSTI_Indore25_26\\ML\\Datasets\\svm_cats_dogs_overlap.csv", index=False)

print("Dataset created: svm_cats_dogs_overlap.csv")
df.head()
```
