```python
import pandas as pd
import numpy as np

np.random.seed(42)

rows = 1000

weight = np.random.uniform(40, 110, rows)

# Linear relation with noise
water_intake = 1 + 0.03 * weight + np.random.normal(0, 0.3, rows)

df2 = pd.DataFrame({
    "Weight_kg": weight.round(1),
    "Daily_Water_Intake_Liters": water_intake.round(2)
})

df2.to_csv("D:\\EduNet\\NSTI_Indore25_26\\ML\\Datasets\\weight_water.csv", index=False)

print("Dataset created: linear_regression_weight_water.csv")
df2.head()
```
