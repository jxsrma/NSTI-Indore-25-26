```python
# Sharpening Filters   
 
import cv2
import matplotlib.pyplot as plt
 
# Read image 
img = cv2.imread(path) 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)    
 
# convert to GrayScal-image 
gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
 
# Laplacian  
laplacian = cv2.Laplacian(gray,cv2.CV_64F)
 
plt.imshow(laplacian, cmap='gray') 
plt.title("Laplacian Filter")
plt.axis('off')
plt.show()
```

```python
# Sobel Filter  
 
import cv2
import matplotlib.pyplot as plt
 
# Read image 
img = cv2.imread(path) 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)    
 
# convert to GrayScal-image 
gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
 
 
# sobel X - vertical edges 
sobelx = cv2.Sobel(gray, cv2.CV_64F, 1,0, ksize=5)
 
# sobel Y - horizontal edges 
sobely =  cv2.Sobel(gray, cv2.CV_64F, 0,1, ksize=5)  
 
plt.imshow(sobelx, cmap='gray')
plt.title("Sobel X")
plt.axis('off')
plt.show()
```

```python
# Harris Corner Detection 

import cv2

import matplotlib.pyplot as plt 

import numpy as np
 
# Read image 

img = cv2.imread("C:\\Users\\Jash\\Downloads\\bmax-blue-ball-pen.jpg") 

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
 
# convert to GrayScal-image 

gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)  
 
# Convert to float32

gray = np.float32(gray)
 
# apply harris corner  

corners = cv2.cornerHarris(gray, 11,11,0.01) #
 
corners = cv2.dilate(corners, None)
 
# corners in red color  

img_rgb[corners > 0.01 * corners.max()] = [255,0,0]
 
plt.imshow(img_rgb)

plt.title("Harris Corners")

plt.axis('off')

plt.show()
 
```

```python
# Shi–Tomasi Corner Detection
 
import cv2
import matplotlib.pyplot as plt 
import numpy as np
 
# Read image 
img = cv2.imread("C:\\Users\\Jash\\Downloads\\bmax-blue-ball-pen.jpg") # BGR 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # RGB  
 
# convert to GrayScal-image 
gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)  # RGB - GRAY  
 
# Corner detection 
corners = cv2.goodFeaturesToTrack(gray,
                                  maxCorners=100,
                                  qualityLevel=0.01,
                                  minDistance=10) 
# convert to integer 
corners = np.int32(corners)
 
# drawing corners 
for corner in corners:
    x,y = corner.ravel() 
    cv2.circle(img_rgb, (x,y), 15, (255,0,0, -1))
 
plt.imshow(img_rgb)
plt.title("Shi-Tomasi Corners")
plt.axis('off')
plt.show() 
```

```python
# Image Alignment and Feature Matching  
 
import cv2
import numpy as np
import matplotlib.pyplot as plt
 
img1 = cv2.imread("C:\\Users\\Jash\\Downloads\\bmax-blue-ball-pen.jpg")
img2 = cv2.imread('C:\\Users\\Jash\\Downloads\\istockphoto-498609933-612x612.jpg')
 
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
 
# Detect ORB Features  - ORB (Oriented FAST and Rotated BRIEF) 
orb = cv2.ORB_create(500) 
kp1, des1 = orb.detectAndCompute(gray1, None) 
kp2, des2 = orb.detectAndCompute(gray2, None)
 
# Match Features 
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True) 
matches = bf.match(des1, des2)
 
matches = sorted(matches, key = lambda x : x.distance)
 
# draw matches 
matched_img = cv2.drawMatches(img1, kp1, img2, kp2, matches[:50], None) 
plt.imshow(cv2.cvtColor(matched_img, cv2.COLOR_BGR2RGB)) 
plt.title("Feature Matching")
plt.axis('off')
plt.show()

# corner. Ravel
```
