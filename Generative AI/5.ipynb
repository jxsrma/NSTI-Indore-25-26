## Stable Diffusion
Stable Diffusion is a text-to-image generative AI model developed by Stability AI in collaboration with academic and open-source partners. It uses a latent diffusion model (LDM) to generate high-quality, realistic images from text prompts, offering a powerful and efficient alternative to earlier, more resource-intensive generative methods.

### Prerequisites
- diffusers
- transformers
- accelerate
- safetensors

Importing:
- StableDiffusionPipleine from diffusers
- torch

```python
from diffusers import StableDiffusionPipeline
```

```python
import torch
```

Using Stable Diffusion model: runwayml/stable-diffusion-v1-5
- Implementing it with StableDiffusionPipeline
- torch_dtype=torch.float16 - Data type selection for Torch
- And mapping it to use "CPU"

```python
model = "runwayml/stable-diffusion-v1-5"
```

```python
sd = StableDiffusionPipeline.from_pretrained(model,
                             torch_dtype = torch.float32).to("cpu")
```

Write a Creative Prompt

```python
prompt = "A Bear in desert"
```

Pass the prompt as and Argument to the model

```python
image = sd(prompt)
```

Access the image using image.images[0]

```python
image.images[0]
```
