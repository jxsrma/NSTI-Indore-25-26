```python
#NER
from transformers import pipeline
ner = pipeline("ner", 
               aggregation_strategy="simple"
               )
text = "Elon Musk founded SpaceX in California."
result = ner(text)
print(result)
# for i in range(len(result)):
#     print(result[i]['entity_group'], ":",result[i]['word'] )
```
