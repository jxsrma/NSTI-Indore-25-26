```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Dataset
data = [45, 60, 40, 55, 70, 65, 50, 45, 90, 85, 60, 55, 50, 45, 70, 
        75, 65, 50, 55, 60, 80, 85, 75, 60, 55, 45, 70, 65, 60, 55]

# Create DataFrame
df = pd.DataFrame(data, columns=['Spending'])
df.head()
```

```python
mean_val = df['Spending'].mean()
median_val = df['Spending'].median()
mode_val = df['Spending'].mode().tolist()

print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")
```

```python
range_val = df['Spending'].max() - df['Spending'].min()
std_dev = df['Spending'].std()

print(f"Range: {range_val}")
print(f"Standard Deviation: {std_dev:.2f}")
```

```python
# Define bins from 40 to 100 with step 10
bins = np.arange(40, 101, 10) 

freq_table = pd.cut(df['Spending'], bins=bins, right=False).value_counts().sort_index().reset_index()
freq_table.columns = ['Interval', 'Frequency']

freq_table
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Spending'], bins=bins, edgecolor='black', color='skyblue')
plt.xlabel('Spending')
plt.ylabel('Frequency')
plt.show()
```

```python
Q1 = df['Spending'].quantile(0.25)
Q3 = df['Spending'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Spending'] < lower_bound) | (df['Spending'] > upper_bound)]

print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
print("Outliers found:")
print(outliers if not outliers.empty else "None")
```
