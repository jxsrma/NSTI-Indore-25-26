```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset of 20 students' heights (in cm)
heights = [155, 160, 162, 165, 167, 168, 170, 170, 172, 174, 
           175, 176, 178, 180, 182, 185, 158, 161, 163, 169]

# Create a DataFrame
df_heights = pd.DataFrame(heights, columns=['Height'])
df_heights.head()
```

```python
Q1 = df_heights['Height'].quantile(0.25)
Q2 = df_heights['Height'].quantile(0.50) # Median
Q3 = df_heights['Height'].quantile(0.75)
IQR = Q3 - Q1

print(f"Q1: {Q1}")
print(f"Q2: {Q2}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
```

```python
plt.boxplot(df_heights['Height'] )
plt.title('Boxplot of Student Heights')
plt.xlabel('Height')
plt.show()
```

```python
mean_h = df_heights['Height'].mean()
std_h = df_heights['Height'].std()
cv = (std_h / mean_h) * 100

print(f"Mean Height: {mean_h:.2f} cm")
print(f"Standard Deviation: {std_h:.2f}")
print(f"Coefficient of Variation: {cv:.2f}%")

# Comment on consistency
if cv < 10:
    print("Comment: The heights are highly consistent (low variability).")
else:
    print("Comment: The heights show moderate to high variability.")
```

```python
def stem_and_leaf(data):
    stems = {}
    for x in sorted(data):
        stem = x // 10
        leaf = x % 10
        if stem not in stems:
            stems[stem] = []
        stems[stem].append(leaf)
    
    print("Stem | Leaf")
    print("-----------")
    for stem, leaves in sorted(stems.items()):
        print(f"{stem:4} | {' '.join(map(str, leaves))}")

stem_and_leaf(heights)
```
