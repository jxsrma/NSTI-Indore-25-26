```python
import pandas as pd
```

```python
df = pd.read_csv("D:\\EduNet\\NSTI_Indore25_26\\ML\\Datasets\\svm_rats_elephants.csv")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x="Height_cm", y="Weight_kg", hue="Animal")
plt.title("Rat vs Elephant (Easy Data)")
plt.show()
```

```python
df = df[["Height_cm","Weight_kg","Animal"]]
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
df['Animal_Encoded'] = LE.fit_transform(df['Animal'])
df
```

```python
df = df[["Height_cm","Weight_kg","Animal_Encoded"]]
df
```

```python
X = df[['Height_cm','Weight_kg']]
y = df['Animal_Encoded']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()

X = ss.fit_transform(X)
X
```

```python
X = pd.DataFrame(X,columns=["H","W"])
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score,confusion_matrix
```

```python
print("accuracy_score:: ",accuracy_score(y_test,y_pred))
print("precision_score:: ",precision_score(y_test,y_pred))
print("recall_score:: ",recall_score(y_test,y_pred))
print("f1_score:: ",f1_score(y_test,y_pred))
print("confusion_matrix:: ",confusion_matrix(y_test,y_pred))
```
