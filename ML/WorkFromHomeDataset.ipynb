````python
import pandas as pd
````

````python
df = pd.read_csv("D:\\EduNet\\NSTI_Indore25_26\\ML\\Datasets\\work_from_home_burnout_dataset.csv")
df
````

````python
df['day_type'].unique()
````

````python
def daysToNumber(day):
    if day == "Weekday":
        return 0
    else:
        return 1
````

````python
df['day_type'] = df['day_type'].apply(daysToNumber)
````

````python
df['burnout_risk'].unique()
````

````python
def riskToNumber(risk):
    if risk == 'Low':
        return 0
    elif risk =='Medium':
        return 1
    else:
        return 2
````

````python
df['burnout_risk'] = df['burnout_risk'].apply(riskToNumber)
df
````

````python
df['burnout_risk'].unique()
````

````python
df.columns
````

````python
X = df[['day_type', 'work_hours', 'screen_time_hours',
       'meetings_count', 'breaks_taken', 'after_hours_work', 'sleep_hours',
        'task_completion_rate']]
y = df['burnout_score']
````

````python
from sklearn.model_selection import train_test_split
````

````python
xtrain,xtest,ytrain,ytest = train_test_split(X,y)
````

````python
from sklearn.linear_model import LinearRegression
````

````python
model = LinearRegression()
model.fit(xtrain,ytrain)
````

````python
ypred = model.predict(xtest)
````

````python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
````

````python
print(mean_absolute_error(ytest,ypred))
print(mean_squared_error(ytest,ypred))
print(root_mean_squared_error(ytest,ypred))
print(r2_score(ytest,ypred))
````
