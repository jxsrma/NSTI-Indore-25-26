```python
import pandas as pd
```

```python
df = pd.read_csv('D:\\EduNet\\NSTI_Indore25_26\\ML\\Datasets\\rainfall.csv')
df
```

```python
X = df.drop(['date','weather_condition'],axis=1)
X
y = df['weather_condition']
```

```python
y.unique()
```

```python
def encodeRainy(weather):
    if weather == 'Rainy':
        return 1
    else:
        return 0
```

```python
y = y.apply(lambda x: encodeRainy(x))
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score,confusion_matrix
print("accuracy_score:: ",accuracy_score(ytest,ypred))
print("precision_score:: ",precision_score(ytest,ypred))
print("recall_score:: ",recall_score(ytest,ypred))
print("f1_score:: ",f1_score(ytest,ypred))
print("confusion_matrix:: ",confusion_matrix(ytest,ypred))
```
