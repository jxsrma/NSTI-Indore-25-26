```python
import pandas as pd
```

```python
df = pd.read_csv("D:\\EduNet\\NSTI_Indore25_26\\ML\\test.csv")
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df[['x']], df['y'], test_size=0.3, random_state=200)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)*100
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['x'], df['y'], color='blue', label='Actual Data Points')
plt.plot(X_test, y_pred, color='red', label='Linear Regression Line')
```
