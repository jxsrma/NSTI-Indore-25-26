```python
import pandas as pd
```

```python
df = pd.read_csv('D:\\EduNet\\NSTI_Indore25_26\\ML\\Datasets\\Student_Performance.csv')
df
```

```python
df['Extracurricular Activities'] = df['Extracurricular Activities'].replace({
    "Yes":1,
    "No":0
})
df
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['Performance Index'],axis=1)
y = df['Performance Index']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,random_state=42)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(xtrain.shape[1],))
])
```

```python
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)
```

```python

model.fit(xtrain, ytrain, epochs=50, batch_size=32)
```

```python

loss, mae = model.evaluate(xtest, ytest)

print("\nTest MSE:", loss)
print("Test MAE:", mae)
```
