```python
import tensorflow
from tensorflow.keras import layers,Sequential
```

```python
train = tensorflow.keras.utils.image_dataset_from_directory(
    "D:\\EduNet\\NSTI_Indore25_26\\Deep Learning\\Dataset\\Dog and Cat\\train",
    image_size = (150,150)    
)
test = tensorflow.keras.utils.image_dataset_from_directory(
    "D:\\EduNet\\NSTI_Indore25_26\\Deep Learning\\Dataset\\Dog and Cat\\test",
    image_size = (150,150)    
)
```

```python
model = Sequential([
    layers.Rescaling(1./255, input_shape=(150, 150, 3)),

    layers.Conv2D(64,(3,3), activation='relu'),
    layers.MaxPool2D((2, 2)),
    
    layers.Conv2D(32,(3,3), activation='relu'),
    layers.MaxPool2D((2, 2)),
    
    layers.Flatten(),
    layers.Dense(16,activation='relu'),
    layers.Dense(8,activation='relu'),
    layers.Dense(1,activation='sigmoid'),
])
```

```python
model.compile(loss="binary_crossentropy",optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(train, 
          validation_data=test, 
          epochs=10, 
          batch_size=64)
```
