```python
URL = "https://aimicrodegree.org/"
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get(URL)
response.status_code
```

```python
soup = BeautifulSoup(response.content,'html')
soup
```

```python
links = soup.find_all('a',href = True)
linkList =[]
for link in links:
    linkList.append(link['href'])
linkList
```

```python
for i in linkList:
    if i[:5] == 'https':
        pass
    else:
        linkList.remove(i)
        print(i)
linkList
    
```
