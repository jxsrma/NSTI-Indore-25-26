```python
students = [
    {"name": "Aditi",
     "marks": 85},
    {"name": "Rahul",
     "marks": 72},
    {"name": "Simran",
     "marks": 90},
    {"name": "Arjun",
     "marks": 67}
]
```

```python
students[1]['marks']
```

```python
for student in students:
    if student['marks'] >= 80:
        print(student['name'])
```

```python
count = 0
for student in students:
    if student['marks'] < 70:
        count += 1
print(count)
```

```python
def get_average(students):
    sum = 0
    for student in students:
        sum += student['marks']
    avg = sum/len(students)
    return avg


print("The average of class test is: ", get_average(students))
```

```python
def get_topper(students):
    topperMarks = -1
    topperName = ""
    for student in students:

        if topperMarks < student['marks']:
            topperMarks = student['marks']
            topperName = student['name']
    return (topperName, topperMarks)


topper = get_topper(students)

print('Topper Name: ', topper[0])
print('Topper Marks: ', topper[1])
```

```python
import statistics as st

marks = []

for stu in students:
    marks.append(stu['marks'])

# print(marks)
print(st.median(marks))


# 85, 72, 90, 67 -> 67, 72, 85, 90

print((72+85)/2)
```
