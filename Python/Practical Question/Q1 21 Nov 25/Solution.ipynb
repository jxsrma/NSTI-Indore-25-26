```python
numbers = [12, 5, 8, 20, 3, 9]

sum = 0
for number in numbers:
    sum += number
print("Sum:", sum)


average = sum/len(numbers)
print("Average:", average)

print("Maximum:", max(numbers))
print("Minimum:", min(numbers))
```

```python
text = "Python programming is fun"

a = 0
e = 0
i = 0
o = 0
u = 0

all_vowels = []

for char in text:
    if char == 'a':
        a += 1
        all_vowels.append(char)
    elif char == 'e':
        e += 1
        all_vowels.append(char)
    elif char == 'i':
        i += 1
        all_vowels.append(char)
    elif char == 'o':
        o += 1
        all_vowels.append(char)
    elif char == 'u':
        u += 1
        all_vowels.append(char)
    else:
        pass


print(f"a: {a}\ne: {e}\ni: {i}\no: {o}\nu: {u}")
print(a+e+i+o+u)
print(all_vowels)
```

```python
data = [1, 4, 7, 12, 15, 18, 21, 24]

odd = []
even = []

for i in data:
    if i % 2 == 0:
        even.append(i)
    else:
        odd.append(i)

print("Even:", even)
print("Odd:", odd)
```
